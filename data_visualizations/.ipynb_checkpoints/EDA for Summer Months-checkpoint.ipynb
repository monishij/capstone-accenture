```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import numpy as np
```

```python
# insert your file path here 
data = pd.read_csv("~/Documents/GitHub/capstone-accenture/clean_data/fireClimateHumanTable_editted.csv")
```

```python
freqs_by_month = data.groupby(data.time).frequency.apply(func=np.sum).to_frame().reset_index()
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
lp = seaborn.lineplot(x=freqs_by_month.time,y=freqs_by_month.frequency)
lp.set_xticks(['2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01'])
lp.set_title('Number of Wildfires Among all Cities Over Time', {'fontsize': 30})
lp.set_xlabel('Date', {'fontsize': 15})
lp.set_ylabel('Number of Wildfires', {'fontsize': 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=data["total_precipitation"], x=data["frequency"])
bp.set_title('Distribution of Total Precipitation for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('Total Precipitation (m)', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
temperature_F = data["2m_temperature"].apply(lambda temp_k : (temp_k - 273.15)*(9/5) + 32)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=temperature_F, x=data["frequency"])
bp.set_title('Distribution of 2m Temperature for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('2m Temperature (Fahrenheit)', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=data["volumetric_soil_water_layer_1"], x=data["frequency"])
bp.set_title('Distribution of Volumetric Soil Water Layer for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('Volumetric Soil Water Layer', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=data["density"], x=data["frequency"])
bp.set_title('Distribution of County Housing Density for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('County Housing Density', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

## Summer Data Visualizations

```python
# well subset june thorugh august (06-08) as the summer months -> highest frequencies
summer_data = data[(data['time'].str.slice(5,7) == '06') | (data['time'].str.slice(5,7) == '07') | (data['time'].str.slice(5,7) == '08')]
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=summer_data["volumetric_soil_water_layer_1"], x=summer_data["frequency"])
bp.set_title('Distribution of Volumetric Soil Water Layer for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('Volumetric Soil Water Layer', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
summer_temperature_F = summer_data["2m_temperature"].apply(lambda temp_k : (temp_k - 273.15)*(9/5) + 32)
bp = seaborn.boxplot(y=summer_temperature_F, x=summer_data["frequency"])
bp.set_title('Distribution of 2m Temperature for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('2m Temperature (Fahrenheit)', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
bp = seaborn.boxplot(y=summer_data["total_precipitation"], x=summer_data["frequency"])
bp.set_title('Distribution of Total Precipitation for Each Frequency Level', {'fontsize': 20})
bp.set_xlabel('Frequency Level', {'fontsize' : 15})
bp.set_ylabel('Total Precipitation (m)', {'fontsize' : 15})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```
