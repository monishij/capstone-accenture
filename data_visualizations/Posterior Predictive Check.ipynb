```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Reading in the predictions from our carbayes model for 2019, and the randomly picked values from rpois().

```python
predicted = pd.read_csv("~/Documents/GitHub/capstone-accenture/transpose_totalCarBayes.csv")
random = pd.read_csv("~/Documents/GitHub/capstone-accenture/transpose_fake_data.csv")
```

Renaming the columns to their respective abbreviated month strings, since they were previously unlabeled.

```python
predicted = predicted.rename(columns={"V1":"Jan", "V2": "Feb", "V3": "Mar", "V4": "Apr", "V5": "May", "V6": "Jun", "V7": "Jul", "V8": "Aug", "V9": "Sep", "V10": "Oct", "V11": "Nov", "V12": "Dec"})
```

```python
predicted = predicted.rename_axis("Month", axis=1)
```

Removing the index column since it's unnecessary.

```python
predicted = predicted.drop(columns=["Unnamed: 0"])
```

```python
predicted = predicted.drop(5000, axis=0)
```

For each column, there are 5000 randomly drawn Poisson values, we want the middle 95% credible interval to compare against our model predictions. Therefore, we calculate the 2.5% and 97.5% quantiles of the values for each month.

```python
quantiles = dict()
for month in predicted.axes[1]:
    quantiles[month] = [np.quantile(predicted[month], 0.025), np.quantile(predicted[month], 0.975)]
quantiles
```

Saving the lower bound and upper bounds of the intervals for ease later while plotting.

```python
first_qs = list()
for _,qs in quantiles.items():
    first_qs.append(qs[0])

second_qs = list()
for _,qs in quantiles.items():
    second_qs.append(qs[1])
```

We need to melt the dataframe into a two column table for ease later in plotting.

```python
predicted = predicted.melt()
```

We make a strip chart of our model predictions for each month, and make a stackplot of the two lines (lower bound and upper bound) of the credible interval.

```python
plt.subplots(1,1, figsize=(10,10))
sns.stripplot(x=predicted.Month, y=predicted.value, data=predicted, alpha=0.2, jitter=0.05, color="green")

y = np.vstack([first_qs, second_qs])
plt.stackplot(quantiles.keys(), y, alpha=0.1, colors = ['None', "green"])
plt.ylabel('Wildfire Occurence Count')
plt.title('Posterior Predictive Check for Riverside in 2019')
plt.show()
```
