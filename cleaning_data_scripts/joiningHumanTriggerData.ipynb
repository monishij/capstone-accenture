```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql+psycopg2://postgres:saadia99@localhost/capstone-accenture')
```

```python
climateFireFrequency = pd.read_sql('SELECT * FROM climateFireFreq', engine)
```

```python
housing = pd.read_sql('SELECT * FROM housingPopulationCounty', engine)
```

```python
population = pd.read_sql('SELECT * FROM populationEstimateCity', engine)
```

```python
housing = housing.rename(columns={'city':'county'})
```

```python
cleaned_housing = housing.melt(id_vars=["county"], 
        var_name="Year", 
        value_name="Density")
```

```python
cleaned_population = population.melt(id_vars=["city"], 
        var_name="Year", 
        value_name="Population")
```

```python
climateFireFrequency = climateFireFrequency.astype({"yearonlyclimate": int})
cleaned_population = cleaned_population.astype({"Year": int})
cleaned_housing = cleaned_housing.astype({"Year": int})
```

```python
climateFireFrequency = climateFireFrequency.rename(columns={'yearonlyclimate':'Year'})
```
