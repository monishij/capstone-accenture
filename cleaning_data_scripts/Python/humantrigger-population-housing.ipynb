```python
import pandas as pd
```

### Cleaning HousingData to import to PostgreSQL

```python
housing = pd.read_csv('../../../capstone-accenture/raw_data/raw_CountyHousingData2.csv')
```

```python
housing
```

```python
# Renaming housing columns
housing.columns = ('county', '2013', '2014', '2015', '2016', '2017', '2018', '2019')
```

```python
# Dropping NA rows
housing = housing.drop([59,60,61,62,63])  #NAs 
```

```python
# String formatting
housing.county = housing.county.str.replace(", California", "")
housing.county = housing.county.str.replace(".", "")
```

```python
housing["2013"] = housing["2013"].str.replace(",", "")
housing["2014"] = housing["2014"].str.replace(",", "")
housing["2015"] = housing["2015"].str.replace(",", "")
housing["2016"] = housing["2016"].str.replace(",", "")
housing["2017"] = housing["2017"].str.replace(",", "")
housing["2018"] = housing["2018"].str.replace(",", "")
housing["2019"] = housing["2019"].str.replace(",", "")
```

```python
# Dataframe transformation
housing = housing.melt(id_vars=["county"], 
        var_name="Year", 
        value_name="Density")
```

```python
# Dataframe to csv
housing.to_csv("newHousingData.csv", index=False)
```
